### 4. Converting the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix

```python
# list of columns to convert
columns_to_convert = ['Value', 'Wage', 'Release Clause']

# Check the data types before conversion
print('dataypes of the columns after converison :\n',df[columns_to_convert].dtypes)

# defining a function convert_to_float
def convert_to_float(column):
    
    # Check if the value is already a float
    if isinstance(column, float):
        return column
    
    # When the record/entry has "M"(indicates millions) as suffix we need to multiply that value with 1000000
    # When the record/entry has "K" (indicates thousands) as suffix we need to multiply that value with 1000
    # currency symbol in this dataset is '€', we will be removing this also along with the suffix
    if 'M' in column and '€' in column: 
        return float(column.replace('M', '').replace('€', ''))*1000000
    elif 'K' in column and '€' in column:
        return float(column.replace('K', '').replace('€', ''))*1000
    if 'M' in column: 
        return float(column.replace('M', ''))*1000000
    elif 'K' in column:
        return float(column.replace('K', ''))*1000
    elif '€' in column:
        return float(column.replace('€', ''))
    else:
        return float(column)

# The convert_value function is applied to each value in the specified columns using the 'apply' function. 
# This function applies the specified function to each element in the Series.
for column in columns_to_convert:
    df[column] = df[column].apply(convert_to_float)

# Check the data types after conversion
print('dataypes of the columns after converison :\n',df[columns_to_convert].dtypes)

# checking random records from the specified columns
df[columns_to_convert].sample(5)
```

### 5. Convert the column "Joined" into integer data type with keeping only the year.

### 7. The column 'Height' is in inches with a quotation mark, Converting it to float with decimal points.

```python
# defining a function to do the conversion

def convert_height_to_dec(height):
    try:
        # Convert height to string
        height_str = str(height)
        if "'" in height_str:
            feet, inches = height_str.split("'")
            # 1 feet = 12 inches
            total_inches = int(feet) * 12 + int(inches)
            height_in_decimal = total_inches / 12
        else:
            # If the value doesn't contain "'", assume it's already in decimal format
            height_in_decimal = float(height_str)
    except Exception as e:
        # If any error occurs during conversion, return NaN
        height_in_decimal = float('nan')
    return height_in_decimal

# Apply the function to the 'Height' column to convert it to decimal feet
df['Height'] = df['Height'].apply(convert_height_to_dec)
```

```python
# check random records from the column 
df['Height'].sample(5)
```

### 8. The column "Weight" has the suffix as lbs, remove the suffix and convert to float.

```python
def convert_weight_to_float(weight):
    try:
        # Convert weight to string
        weight_str = str(weight)
        # Remove "lbs" suffix and convert to float
        weight_float = float(weight_str.replace("lbs", ""))
        return weight_float
    except ValueError:
        # If the conversion fails, return NaN
        return float('nan')

# Apply the function to the 'Weight' column to remove the suffix and convert to float
df['Weight'] = df['Weight'].apply(convert_weight_to_float)
```

```python
# check ramdom records for the column

df['Weight'].sample(5)
```

### 9. Check for the percentage of missing values and impute them with appropriate imputation techniques.

```python
# Calculate the percentage of missing values in each column
missing_percentage = df.isna().mean() * 100
# Print the percentage of missing values for each column
print("Percentage of missing values in each column:")
print(missing_percentage)
```

- percentage of missing values in almost all columns are negligible and can be imputed using the proper imputation techniques
- However, 'Loaned From' column has too many missing values and is not relevant to the analysis. So, it will be dropped from the dataset.
- The other columns will be differentiated on the basis of type of the variable(numercial or categorical) and if they have missing values or not. Then, the respective imputation techniques will be applied to fill the missing records

```python
# dropping 'Loaned From' column

df.drop('Loaned From', axis=1, inplace = True)

# checking if the columns are dropped or not
removed = True
if 'Loaned From' in df.columns:
    removed = False
        
if all_removed:
    print("The column has been removed.")
else:
    print("The column has not been removed.")
```

```python
# Imputation techniques by filtering columns based on type and presence of null values
# Filter columns with null values
columns_with_null = df.columns[df.isnull().any()]

# Filter null columns based on their data types
numerical_columns = df[columns_with_null].select_dtypes(include=['number']).columns
categorical_columns = df[columns_with_null].select_dtypes(include=['object']).columns

# Convert numerical columns to float64 to allow for decimal values
df[numerical_columns] = df[numerical_columns].astype('float64')

# Impute missing values for numerical columns with mean
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())

# Impute missing values for categorical columns with mode
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])

# Check if any missing values remain
print("Remaining missing values after imputation:")
print(df.isnull().sum())
```

- 'Contract Valid' until column still has missing values. we can separately take this columns and impute it with the most frequent entry

```python
# imputing missing values in 'Contract Valid Until' by replacing them with the most frequent date

most_frequent_contract_end_date = df['Contract Valid Until'].mode().iloc[0]
df['Contract Valid Until'] = df['Contract Valid Until'].fillna(most_frequent_contract_end_date)

# Checking again if any missing values remain
print("Remaining missing values after imputation:")
print(df.isnull().sum())
```

- All missing values have been successfully imputed.

### 21. What is the average wage one can expect to pay for the top 5 players in every position?

- We have already found the top players in each position. we will use that information here

```python
# Group the DataFrame containing top 5 players by Overall rating for each unique position by 'Position'
position_groups = top_players_by_position_df.groupby('Position')

# Initialize an empty dictionary to store the average wages for each position
average_wages = {}

# Iterate over each position group
# 'groupby' generates a tuple the position in the first elemenst, second element is the group of rows filtered
# here we will take 'position' as key and the average wage of the 'group' as value
for position, group in position_groups:
    # Calculate the average wage for these top 5 players in the position
    average_wage = group['Wage'].mean()
    # Store the average wage for this position in the dictionary
    average_wages[position] = average_wage

# Convert the dictionary to a DataFrame for easier visualization
average_wages_df = pd.DataFrame.from_dict(average_wages, orient='index', columns=['Average Wage'])

# Display the DataFrame
print('The average wage one can expect to pay for the top 5 players in every position :\n')
print(average_wages_df)
```
